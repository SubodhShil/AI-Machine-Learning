Matplotlib

```python
# importing matplotlib and numpy library
import matplotlib.pyplot as mplt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
```

Plotting the data

```python
# sine wave
mplt.plot(x, y)
mplt.show()
```

```python
# cosine wave
mplt.plot(x, z)
mplt.show()
```

```python
# Adding title, x-axis and y-axis labels
mplt.plot(x, y)
mplt.xlabel("angle")
mplt.ylabel("sine value")
mplt.title("sine wave")
mplt.show()
```

```python
# Parabola
x = np.linspace(-10, 10, 20)
y = x**2
mplt.plot(x, y, "g*")
mplt.show()
```

```python
x = np.linspace(-5, 5, 50)
mplt.plot(x, np.sin(x), 'g')
mplt.plot(x, np.cos(x), 'r')
mplt.show()
```

Bar plot

```python
fig = mplt.figure()
ax = fig.add_axes([0, 0, 1, 1])
countries = ['USA', 'Russia', 'France', 'Germany', 'Japan']
people = ['330M', '150M', '50M', '60M', '110M']
ax.bar(countries, people)
mplt.xlabel('Countries')
mplt.ylabel('Population')
```
