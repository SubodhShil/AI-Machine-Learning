```python
import pandas as pd
import numpy as np
```

```python
index = 'r1 r2 r3 r4 r5 r6 r7 r8 r9 r10'.split()
columns = 'c1 c2 c3 c4 c5 c6 c7 c8 c9 c10'.split()
```

Using arange make a 2d array (10, 10) of 100 to 0 values.

```python
array_2d = np.matrix(np.arange(100, 0, -1).reshape(10, 10))
array_2d 
```

```python
df = pd.DataFrame(array_2d, index=index, columns=columns)
print(df)
```

df is our first dataframe.
We have columns, c1 to c10, and their corresponding rows, r1 to r10.
Each column is actually a pandas series, sharing a common index (row labels).

â˜ž Let's learn how to Grab data that we need, this is the most important thing we want to learn to move one!

Grab a single column c10

```python
column_10 = df['c10']
print(column_10)
```

Grab multiple columns c4, c2, c9

```python
multi_column = df[['c4', 'c2', 'c9']]
multi_column
```

Add a new column name 'new' and value would be c1 * c10

```python
c1_multiplication_c10 = df['c1'] * df['c10']
c1_multiplication_c10
```

```python
df = df.assign(new=c1_multiplication_c10)
df
```

Delete the column 'new' permanently

```python
del df['new']
df
```

Grabb more than one rows 'r2', 'r1','r3'

```python
grabbed_rows = pd.DataFrame(df, index=['r2', 'r1', 'r3'])
grabbed_rows
```

Grab a sub-set of the dataframe where c1 is less than 50

```python
subset_dataframe = df[df.lt(50).all(axis=1)]
subset_dataframe
```

Grab a sub-set of the dataframe where c3 is equal to 38 and c8 is equal to 33

```python
subset_dataframe2 = df[(df['c3'] == 38) & (df['c8'] == 33)]
subset_dataframe2
```
