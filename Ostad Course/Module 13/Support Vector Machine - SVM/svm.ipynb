```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
```

```python
df = pd.read_csv('Breast_Cancer_Diagnostic.csv')
df.sample(10)
```

Train test split 

```python
X = df.drop('diagnosis', axis=1)
Y = df['diagnosis']
```

```python
Y
```

Approach 1

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

clf = GridSearchCV(SVC(gamma='auto'), {
    'C': [1, 10, 20, 30],
    'kernel': ['rbf', 'linear']
}, cv=5)

clf.fit(X, Y)
```

```python
test = pd.DataFrame(clf.cv_results_)
test
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=2)
```

```python
svm_model = SVC(kernel='linear', C=30, gamma='auto')
```

```python
svm_model.fit(X_train, y_train)
```

```python
svm_model.score(X_test, y_test)
```

Approach 2

```python
from sklearn.model_selection import cross_val_score
```

```python
sv1 = cross_val_score(SVC(kernel='linear', C=1, gamma='auto'), X, Y, cv=5)
sv1.mean()
```

```python
sv2 = cross_val_score(SVC(kernel='rbf', C=1, gamma='auto'), X, Y, cv=5)
sv2.mean()
```
