```python
import pandas as pd
from matplotlib import pyplot as plt
```

**load the csv**

```python
#code here
df = pd.read_csv('titanic_full.csv')
```

```python
df.sample(10)
```

**Check the null value**

```python
df.isnull().sum()
```

**Plot the data**

```python
plt.scatter(df[['Pclass']], df[['Survived']], marker='+', color='red')
```

**Separate the Features a target in x and y**

```python
#code here  
x = df[['Fare']]
y = df[['Survived']]
```

```python
x.head()
```

```python
y.head()
```

**Split the train and test data**

```python
#code here
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.8, random_state = 2)
```

```python
X_train.head()
```

```python
y_train.head()
```

**Make the model Binary logistic regression model**

```python
# code here
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

**Check the model score**

```python
# code here
y_predict = model.predict(X_test)
y_predict
```

```python
model.score(X_test, y_test)
```
