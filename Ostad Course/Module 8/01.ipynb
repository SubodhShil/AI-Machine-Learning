```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(11)
y = x ** 2
print(x, y)
```

```python
plt.scatter(x, y, color='red')
```

```python
plt.plot(x, y, 'r-*')
plt.show()
```

```python
# Adding title in the figure 

plt.title("First plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.plot(x, y, 'g-+')
plt.show()
```

```python
# Control x-axis and y-axis digits numbers without using numpy 
plt.title("Plot 2")
plt.xlabel("X axis")
plt.ylabel("X axis")

plt.xticks(range(0, 11))
plt.yticks(range(0, 110, 5))

a = np.arange(11)
b = a ** 2
plt.plot(a, b, 'r-+')

plt.show()
```

### Multiple plots at a time

```python
# First subplot
x1 = [1, 2, 3, 4]
y1 = [2, 5, 13, 20]
plt.subplot(1, 3, 1)
plt.plot(x1, y1)

# Second subplot
x2 = [1, 2, 3, 4]
y2 = [3, 4, 5, 6]
plt.subplot(1, 3, 2)
plt.plot(x2, y2, 'r--')

# Third subplot
x2 = [1, 2, 3, 4]
y2 = [5, 10, 15, 20]
plt.subplot(1, 3, 3)
plt.plot(x2, y2)
```
