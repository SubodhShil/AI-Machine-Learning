```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model 
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df.head()
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Per Capita Income')
plt.title('Per Capita Income Over Years')
plt.scatter(df['year'], df['per capita income (US$)'], color='green', marker='+')
plt.show()
```

```python
linear_reg = linear_model.LinearRegression()
linear_reg.fit(df[['year']], df[['per capita income (US$)']])
```

```python
linear_reg.predict([[2018]])
```

```python
%matplotlib inline 
plt.xlabel('Year')
plt.ylabel('Per Capita Income')
plt.scatter(df['year'], df['per capita income (US$)'], color='green', marker='+')
plt.plot(df['year'], linear_reg.predict(df[['year']]), color='blue')
```
