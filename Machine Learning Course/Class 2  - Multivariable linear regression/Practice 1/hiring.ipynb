```python
import numpy as np
import pandas as pd 
from sklearn import linear_model
```

```python
df = pd.read_csv('hiring.csv')
df
```

```python
# check if there any null value available in the dataframe
print(df.isnull().sum())
```

Handling experience column

```python
%pip install word2number
```

```python
from word2number import w2n
# handle NaN values
df.experience = df.experience.fillna('zero')
df.experience
```

```python
new_experience_col = df['experience'].apply(w2n.word_to_num)
new_experience_col
```

```python
# drop old 'experience' column
df.drop(columns=['experience'], inplace=True)
df.insert(loc=1, column='experience', value=new_experience_col)
```

```python
# Fill NaN values in 'test_score(out of 10)' and 'interview_score(out of 10)' columns with the median value
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(df['test_score(out of 10)'].median())
df['interview_score(out of 10)'] = df['interview_score(out of 10)'].fillna(df['interview_score(out of 10)'].median())
```

```python
df
```

```python
linear_regression = linear_model.LinearRegression()
linear_regression.fit(df[['test_score(out of 10)', 'experience', 'interview_score(out of 10)']], df[['salary($)']])
```

```python
linear_regression.predict([[7.0, 10, 8]])
```
