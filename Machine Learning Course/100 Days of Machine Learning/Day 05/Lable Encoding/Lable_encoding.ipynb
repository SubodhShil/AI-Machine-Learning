# **Label encoding**



```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

### **Lable Encoding of Breast Cancer Dataset**

```python
breast_cancer_data = pd.read_csv("/content/breast_cancer_data.csv")
```

```python
breast_cancer_data.head()
```

```python
# checking which unique datas are present
breast_cancer_data['diagnosis'].unique()
```

```python
# counting how many unique values are present
print(breast_cancer_data['diagnosis'].value_counts())
```

### Encoding process

```python
# load the encoder method
label_encode = LabelEncoder()
```

```python
labels = label_encode.fit_transform(breast_cancer_data['diagnosis'])
```

```python
# appending labels as column in the dataframe
breast_cancer_data['target'] = labels
```

```python
breast_cancer_data.head()
```

```python
# 0 -> Benign, 1 -> Malignant
breast_cancer_data.target.value_counts()
```

### **Lable Encoding of Iris Data**

```python
iris_data = pd.read_csv("/content/iris_data.csv")
```

```python
iris_data.head()
```

```python
iris_data.Species.value_counts()
```

```python
iris_data_label_encoding = LabelEncoder()
```

```python
new_label = iris_data_label_encoding.fit_transform(iris_data.Species)
```

```python
iris_data['result'] = new_label
```

```python
iris_data.result.value_counts()

'''
Here,

Iris-setosa        ->   0
Iris-versicolor    ->   1
Iris-virginica     ->   2
'''
```
